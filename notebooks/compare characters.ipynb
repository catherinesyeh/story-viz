```python
import json

story_name = "odyssey"
with open(f'/Users/wangxilin/story-viz/src/data/{story_name}-new.json', 'r') as file:
    data = json.load(file)

all_characters = [character['name'] for character in data.get('characters', [])]
set_ribbons = set(all_characters)

all_characters_sparknotes = [
    "Odysseus",
    "Telemachus",
    "Penelope",
    "Athena",
    "Calypso",
    "Circe",
    "Poseidon",
    "Zeus",
    "Antinous",
    "Eurymachus",
    "Amphinomus",
    "Eumaeus",
    "Eurycleia",
    "Melanthius",
    "Melantho",
    "Polyphemus",
    "Laertes",
    "Tiresias",
    "Nestor",
    "Menelaus",
    "Helen",
    "Agamemnon",
    "Achilles",
    "Nausicaa",
    "Alcinous",
    "Arete"
]

set_sparknotes = set(all_characters_sparknotes)

# Find differences
only_in_sparknotes = set_sparknotes - set_ribbons
only_in_ribbons = set_ribbons - set_sparknotes
both = set_sparknotes & set_ribbons

# Convert to sorted lists for better readability
only_in_sparknotes_list = sorted(list(only_in_sparknotes))
only_in_ribbons_list = sorted(list(only_in_ribbons))
both_list = sorted(both)

# Create result dictionary
result = {
    "only_in_sparknotes": {
        "characters": only_in_sparknotes_list,
        "count": len(only_in_sparknotes_list)
    },
    "only_in_ribbons": {
        "characters": only_in_ribbons_list,
        "count": len(only_in_ribbons_list)
    },    
    "both":{
        "characters": both_list,
        "count": len(both_list)
    }
}

# Save to JSON
with open(f'/Users/wangxilin/story-viz/notebooks/comparisons/compare characters/character-diff-{story_name}.json', 'w') as outfile:
    json.dump(result, outfile, indent=4)

print("Comparison saved to character-diff.json")
```
