```python
import json

story_name = "odyssey"
with open(f'/Users/wangxilin/story-viz/src/data/{story_name}-new-themes.json', 'r') as file:
    data = json.load(file)

all_themes = [character['name'] for character in data.get('characters', [])]
set_ribbons = set(all_themes)

all_themes_sparknotes = [
    "The Power of Cunning over Strength",
    "The Pitfalls of Temptation",
    "Divine Justice",
    "Homecoming",
    "Cunning"   
]

set_sparknotes = set(all_themes_sparknotes)

# Find differences
only_in_sparknotes = set_sparknotes - set_ribbons
only_in_ribbons = set_ribbons - set_sparknotes
both = set_ribbons & set_sparknotes

# Convert to sorted lists for better readability
only_in_sparknotes_list = sorted(list(only_in_sparknotes))
only_in_ribbons_list = sorted(list(only_in_ribbons))
both_list = sorted(list(both))

# Create result dictionary
result = {
    "only_in_sparknotes": {
        "themes": only_in_sparknotes_list,
        "count": len(only_in_sparknotes_list)
    },
    "only_in_ribbons": {
        "themes": only_in_ribbons_list,
        "count": len(only_in_ribbons_list)
    },
    "both":{
        "themes": both_list,
        "count": len(both_list)
    }
}

# Save to JSON
with open(f'/Users/wangxilin/story-viz/notebooks/comparisons/compare themes/theme-diff-{story_name}.json', 'w') as outfile:
    json.dump(result, outfile, indent=4)

print("Comparison saved to theme-diff.json")
```
